```python
import os
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 資料路徑
DATASET_PATH  = '/home/cslab/Documents/NASA/'

data_dir = '2nd_test/'
```

```python
raw_data = pd.DataFrame()
for filename in os.listdir(DATASET_PATH + data_dir):
#     print(filename)
    dataset = pd.read_csv(os.path.join(DATASET_PATH+data_dir, filename), sep='\t')
    dataset_mean = np.array(dataset.mean())
    dataset = pd.DataFrame(dataset_mean.reshape(1,4))
    dataset.index = [filename]
    raw_data = raw_data.append(dataset)
    
raw_data.columns = ['Bearing 1', 'Bearing 2', 'Bearing 3', 'Bearing 4']
```

```python
raw_data.index = pd.to_datetime(raw_data.index, format='%Y.%m.%d.%H.%M.%S')
```

```python
raw_data = raw_data.sort_index()
```

```python
raw_data.to_csv('test_dataset_BearingTest_2.csv')
```
